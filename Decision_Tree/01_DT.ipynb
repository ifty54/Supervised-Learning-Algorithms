```python
import pandas as pd
```

```python
path = "/content/salaries.csv"
df = pd.read_csv(path)
```

```python
df.head()
```

```python
features = df.drop('salary_more_then_100k', axis = 'columns')
```

```python
target = df["salary_more_then_100k"]
```

```python
features
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoded_company = LabelEncoder()
encoded_job = LabelEncoder()
encoded_degree = LabelEncoder()
```

```python
features["company_n"] = encoded_company.fit_transform(features["company"])
features["job_n"] = encoded_job.fit_transform(features["job"])
features["degree_n"] = encoded_degree.fit_transform(features["degree"])
```

```python
features
```

```python
new_features = features.drop(['company','job','degree'],axis='columns')
new_features
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(new_features, target)
```

```python
model.score(new_features, target)
```
